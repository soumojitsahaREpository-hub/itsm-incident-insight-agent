## Stage 8.1 : Incident Lifecycle Understanding — How Tickets Move Over Time

````python
df_lifecycle = df.copy()

column_change_summary = []

for col in df_lifecycle.columns:
    incidents_where_changed = (
        df_lifecycle.groupby("number")[col]
                    .nunique(dropna=False)
                    .gt(1)
                    .sum()
    )

    column_change_summary.append({
        "column": col,
        "incidents_where_column_changed": incidents_where_changed,
        "%_of_incidents_changed": round(
            incidents_where_changed / df_lifecycle["number"].nunique() * 100,
            2
        )
    })

column_change_summary_df = (
    pd.DataFrame(column_change_summary)
    .sort_values("incidents_where_column_changed", ascending=False)
)

column_change_summary_df

````

## Stage 8.2 : Validate cumulative counters

````python
df_lifecycle["sys_updated_at_parsed"] = pd.to_datetime(df_lifecycle["sys_updated_at"], errors="coerce", dayfirst=True)

def has_decreasing_counter(group, counter_col):
    group = group.sort_values(["sys_updated_at_parsed", "sys_mod_count"])
    return (group[counter_col].diff().dropna() < 0).any()


counter_monotonicity_results = []

for counter_col in ["reassignment_count", "reopen_count", "sys_mod_count"]:
    incidents_with_decrease = (
        df_lifecycle
        .groupby("number")
        .apply(lambda group: has_decreasing_counter(group, counter_col))
        .sum()
    )

    counter_monotonicity_results.append({
        "counter_column": counter_col,
        "incidents_with_decreasing_counter": incidents_with_decrease
    })

pd.DataFrame(counter_monotonicity_results)

````

# Stage 8 Conclusion: 


### Aggregation Rule Planning

The raw dataset is event-level/history-level, meaning one incident ticket can appear across multiple rows as it changes over time.

For future incident-level analysis, the data will later need to be converted into one row per incident. This stage does not create that final dataset. It only defines the aggregation rules that should be used later in Data cleaning notebook or the agent-ready preparation step.

Aggregation is needed because raw history rows should not be counted directly as incidents. For example, one incident may have many updates, reassignment changes, state changes, and system modifications. Without aggregation, incident volume, SLA distribution, assignment group workload, reopen counts, and reassignment analysis could be misleading.

### Planned Aggregation Rules

| Column               | Aggregation Rule                                 | Reason                                                                                                                         |
| -------------------- | ------------------------------------------------ | ------------------------------------------------------------------------------------------------------------------------------ |
| `number`             | `group key`                                      | Identifies the unique incident ticket. All event/history rows with the same number belong to the same incident.                |
| `opened_at`          | `earliest valid value`                           | Represents when the ticket was first opened. This is the start point of the incident lifecycle.                                |
| `sys_updated_at`     | `latest valid value`                             | Represents the latest update timestamp for the incident history. Useful for identifying the final/latest row.                  |
| `resolved_at`        | `latest valid value`                             | Represents when the incident was marked resolved. The latest valid value should represent the final resolved timestamp.        |
| `closed_at`          | `latest valid value`                             | Represents when the incident was officially closed. The latest valid value should represent the final closure timestamp.       |
| `incident_state`     | `value from latest sys_updated_at row`           | Represents the latest known state of the incident, such as Active, Resolved, or Closed.                                        |
| `active`             | `value from latest sys_updated_at row`           | Represents whether the incident was active in the latest known ticket state.                                                   |
| `made_sla`           | `value from latest sys_updated_at row`           | Represents the latest known SLA status for the incident.                                                                       |
| `priority`           | `value from latest sys_updated_at row`           | Represents the latest known priority after all ticket updates.                                                                 |

### Assumption Validation Table
